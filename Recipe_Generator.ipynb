```python
df = pd.read_csv("RAW_recipes.csv") 
df.head()
```

```python
df.info()  
df.head() 
```

```python
df = df.dropna(subset=['ingredients'])  # Drop rows with missing ingredients
df.reset_index(drop=True, inplace=True)  # Reset index
```

```python
import ast

# Convert string to actual list
df['ingredients'] = df['ingredients'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Normalize text: convert to lowercase and strip spaces
df['ingredients'] = df['ingredients'].apply(lambda x: [i.lower().strip() for i in x])
```

```python
df.dropna(subset=['ingredients'], inplace=True)  # Remove empty ingredient lists
df.drop_duplicates(subset=['name'], inplace=True)  # Drop duplicate recipes
df.reset_index(drop=True, inplace=True)  # Reset index
```

```python
df.dropna(subset=['name'], inplace=True)
df.reset_index(drop=True, inplace=True)  # Reset index after dropping
```

```python
print(df.isnull().sum())  # Should show 0 missing values in 'name'
```

```python
df['ingredients_text'] = df['ingredients'].apply(lambda x: ' '.join(x))  # Join list into a string
df[['name', 'ingredients_text']].head()  # Check output
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Create TF-IDF model
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df['ingredients_text'])  # Transform ingredients into vectors

print(tfidf_matrix.shape)  # Output should be (num_recipes, num_unique_ingredients)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend_recipes(user_ingredients, top_n=5):
    # Convert user input into a list & preprocess
    user_ingredients = [i.lower().strip() for i in user_ingredients]
    user_input_text = ' '.join(user_ingredients)  # Convert list to string
    
    # Convert input to TF-IDF vector
    user_tfidf = vectorizer.transform([user_input_text])
    
    # Compute cosine similarity
    similarities = cosine_similarity(user_tfidf, tfidf_matrix)
    
    # Get top N matching recipes
    top_indices = similarities.argsort()[0][-top_n:][::-1]  # Get top N indices
    
    # Return matching recipe names
    return df.iloc[top_indices][['name', 'ingredients']]

# Example test
user_input = ["tomato", "cheese", "bread"]  # Example fridge items
print(recommend_recipes(user_input))

```
